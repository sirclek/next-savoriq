```python
chemical_id_map = {chem["name"]: chem["id"] for chem in data["chemicals"]}

# Function to convert compounds into a 20-element array
def convert_compounds_to_array(whiskey):
    compounds_array = [0] * 20
    for compound, value in whiskey["compounds"].items():
        if compound in chemical_id_map:
            index = chemical_id_map[compound] - 1  # Offset by 1 for 0-based index
            compounds_array[index] = value
    return compounds_array

# Convert compounds for each whiskey
for whiskey in data["whiskeys"]:
    whiskey["compounds_array"] = convert_compounds_to_array(whiskey)

class CompactArrayEncoder(json.JSONEncoder):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def iterencode(self, obj):
        if isinstance(obj, list) and all(isinstance(i, int) for i in obj):
            # Ensures lists of integers (like compounds_array) stay on one line
            return '[' + ', '.join(map(str, obj)) + ']'
        else:
            return super().iterencode(obj)

# Write to JSON file with custom encoder
output_file = 'whiskey_data_with_compounds.json'
with open(output_file, 'w') as f:
    json.dump(data, f, cls=CompactArrayEncoder, indent=4)

print(f"Data with compounds array has been written to {output_file}")
```
